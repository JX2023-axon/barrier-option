# Heston Model

$$
dS_t = \mu S_t\,dt + \sqrt{v_t}S_t\,dW_{1t} \tag{1}
$$
$$
dv_t = \kappa(\theta-v_t)\,dt + \sigma\sqrt{v_t}\,dW_{2t} \tag{2}
$$
$$
dW_{1t}\,dW_{2t}=ρ\,d_t \tag{3}
$$

#### 标的价格、方差的随机过程
公式1：S是标的价格。μ代表收益率，v代表标的价格方差。第一项代表标的随时间迁移。第二项代表波动的
维纳过程。  
公式2：方差与标的价格一样为随机过程，补充Black-Scholes-Merton公式中对波动率为常数，或波动率表
面中将波动率作为独立变量的不足。θ代表预期方差，从而公式2的第一项保证方差的均值回归，第二项代表
维纳过程。公式2来自奥恩斯坦-乌伦贝克假设[Ornstein-Uhlenbeck Assumption](#从奥恩斯坦-乌伦贝克假设到赫斯顿模型)。  
公式3：方差与标的价格的维纳过程符合相关性为ρ。  

#### 构建无风险套利组合
为获得衍生品价格与标的价格、波动率、时间相关的关系，下面构建对冲的无风险套利投资组合。  
假设衍生品为$V(S,v,t)$，$S$和$v$拥有两个维纳过程$W_{1t},W_{2t}$，需要两个衍生品$U,V$进行组合
构建。假设组合包含：  
$1\,V$  
$Δ\,S$  
$ϕ\,U$  
则投资组合$Π$为：  
$Π\,=\,V\,+\,Δ\,S\,+\,ϕ\,U$  
$dΠ\,=\,dV\,+\,Δ\,dS\,+\,ϕ\,dU$  
下面对$U,V$的随机过程通过伊藤引理进行推导。

$$
dV=
\left[
\frac{\partial V}{\partial t}
+\frac{1}{2}vS^2\frac{\partial^2 V}{\partial S^2}
+\rho\sigma vS\frac{\partial^2 V}{\partial S\,\partial v}
+\frac{1}{2}\sigma^2 v\frac{\partial^2 V}{\partial v^2}
\right]dt
+\frac{\partial V}{\partial S}\,dS
+\frac{\partial V}{\partial v}\,dv
\tag{4}
$$  

$$
d\Pi
=
\left(\,[\cdots]_V+\phi[\cdots]_U\,\right)dt
+
\left(\frac{\partial V}{\partial S}+\phi\frac{\partial U}{\partial S}+\Delta\right)dS
+
\left(\frac{\partial V}{\partial v}+\phi\frac{\partial U}{\partial v}\right)dv
\tag{5}
$$

其中$Δ,ϕ$是用于构建无风险投资组合，因此根据公式（5）$dS,dv$前因子应为0：  
消除方差风险 $dv$:
$$
\frac{\partial V}{\partial v}
+\phi\,\frac{\partial U}{\partial v}
=0
\;\Longrightarrow\;
\phi
=
-\frac{\partial V/\partial v}{\partial U/\partial v}\tag{6}
$$

消除价格风险 $dS$:
$$
\frac{\partial V}{\partial S}
+\phi\,\frac{\partial U}{\partial S}
+\Delta
=0
\;\Longrightarrow\;
\Delta
=
-\frac{\partial V}{\partial S}
-\phi\,\frac{\partial U}{\partial S}\tag{7}
$$

#### 套利组合的伊藤过程与无风险收益率
进一步的，因为投资组合为无风险套利组合，其收益率仅与无风险收益率$r$和时间$t$相关
$$
d\Pi = r\Pi\,dt = r\left(V+\Delta S+\phi U\right)dt
\tag{8}
$$

在公式（5）中其收益率为
$$
\left[
\frac{\partial V}{\partial t}
+\frac{1}{2}vS^2\frac{\partial^2 V}{\partial S^2}
+\rho\sigma vS\frac{\partial^2 V}{\partial S\,\partial v}
+\frac{1}{2}\sigma^2 v\frac{\partial^2 V}{\partial v^2}
\right]
+\phi\left[
\frac{\partial U}{\partial t}
+\cdots
\right]
$$

从而得到其等于$r\,(V\,+\,Δ\,S\,+\,ϕ\,U)$  
$$
\begin{gathered}
\frac{\partial V}{\partial t}
+\frac{1}{2}vS^2\frac{\partial^2 V}{\partial S^2}
+\rho\sigma vS\frac{\partial^2 V}{\partial S\,\partial v}
+\frac{1}{2}\sigma^2 v\frac{\partial^2 V}{\partial v^2}
+\phi\left[\frac{\partial U}{\partial t}+\cdots\right]
=\\
r\left(V+\Delta S+\phi U\right)
\end{gathered}\tag{9}
$$
从公式（6）和（7）代入$Δ,ϕ$并化简  

$$
\begin{aligned}
\frac{
\frac{\partial V}{\partial t}
+ \frac{1}{2} v S^2 \frac{\partial^2 V}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 V}{\partial S \partial v}
+ \frac{1}{2} \sigma^2 v \frac{\partial^2 V}{\partial v^2}
- rV
+ rS \frac{\partial V}{\partial S}
}{
\frac{\partial V}{\partial v}
}
=\\
\frac{
\frac{\partial U}{\partial t}
+ \frac{1}{2} v S^2 \frac{\partial^2 U}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 U}{\partial S \partial v}
+ \frac{1}{2} \sigma^2 v \frac{\partial^2 U}{\partial v^2}
- rU
+ rS \frac{\partial U}{\partial S}
}{
\frac{\partial U}{\partial v}
}
\tag{10}
\end{aligned}
$$


注意到公式（10）左右两侧形式相同，且每侧仅与一个衍生品相关。同时，任意衍生品$U,V$适用于
公式（10），其应为一个普适的函数$f(S,v,t)$。将${\frac{\partial U}{\partial v}}$乘到
$f(S,v,t)$并根据[费曼-卡兹定理](#费曼-卡茨定理feynman-kac-theorem)，$f(S,v,t)$因为是衍生品价格对$v$的一阶偏导，因此$f(S,v,t)$代表
$v$的风险中性漂移。


#### $f(S,v,t)$的形式
因为$f(S,v,t)$为方差$v$的风险中性漂移率，可得，
$$
f(S,v,t) = -\kappa(\theta - v) + \lambda(S,v,t)\tag{11}
$$
其中，$\lambda(S,v,t)$是市场对于$v$的风险溢价，$-\kappa(\theta - v)$代表风险市场中的$v$的漂移率。
借鉴布里登Breeden's(1979)基于消费的资产定价模型，和考克斯-英格索尔-罗斯CIR(1985)的一般均衡模型，
$v$的风险溢价与$v$成正比，从而有：
$$
f(S,v,t) = \kappa(\theta - v) - \lambda v\tag{12}
$$
若给定：$\qquad \kappa^* = \kappa + \lambda\,,\,\theta^* = \frac{\kappa \theta}{\kappa + \lambda}$
$$
f(S,v,t) = \kappa^*(\theta^* - v)\tag{13}
$$

#### Heston模型下的PDE
Heston模型结合了
1. 标的价格的随机过程
2. 标的方差的随机过程
3. 两者的相关性

从而得出一般性的 Heston PDE:

$$
\begin{aligned}
\frac{\partial U}{\partial t}
&+ \frac{1}{2} vS^2 \frac{\partial^2 U}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 U}{\partial S \partial v}
+ \frac{1}{2} \sigma^2 v \frac{\partial^2 U}{\partial v^2} \\
&+ rS \frac{\partial U}{\partial S}
+ \bigl[\kappa(\theta - v) - \lambda(S,v,t)\bigr] \frac{\partial U}{\partial v}
- rU = 0
\tag{14}
\end{aligned}
$$

若假定 $v$ 的风险溢价与 $v$ 成正比：

$$
\begin{aligned}
\frac{\partial C}{\partial t}
&+ \frac{1}{2} vS^2 \frac{\partial^2 C}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 C}{\partial S \partial v}
+ \frac{1}{2} \sigma^2 v \frac{\partial^2 C}{\partial v^2} \\
&+ rS \frac{\partial C}{\partial S}
+ \kappa^*(\theta^* - v)\frac{\partial C}{\partial v}
- rC = 0
\tag{15}
\end{aligned}
$$


# 从奥恩斯坦-乌伦贝克假设到赫斯顿模型

从波动率 $h_t$ 的均值回归过程出发，利用伊藤引理推导出方差 $v_t=h_t^2$的随机微分方程。

假设:
$$
dh_t = -\beta h_t\,dt + \delta\,dW_{2t} \tag{S1}
$$

假设$dh_t = a\,dt + b\,dW_{2t}$, 得出:
$$
a(h_t) = -\beta h_t,\qquad b(h_t)=\delta \tag{S2}
$$

定义方差为:
$$
v_t = G(h_t)=h_t^2 \tag{S3}
$$

依据伊藤引理:
$$
\frac{\partial G}{\partial h}=2h_t,\qquad
\frac{\partial^2 G}{\partial h^2}=2,\qquad
\frac{\partial G}{\partial t}=0 \tag{S4}
$$

$$
dG=\left(\frac{\partial G}{\partial h}a+\frac{\partial G}{\partial t}
+\frac{1}{2}\frac{\partial^2 G}{\partial h^2}b^2\right)dt
+\frac{\partial G}{\partial h}b\,dW_{2t} \tag{S5}
$$

代入$a,b$和导数:
$$
dv_t=\left((2h_t)(-\beta h_t)+\frac{1}{2}(2)\delta^2\right)dt
+(2h_t)\delta\,dW_{2t} \tag{S6}
$$

化简:
$$
dv_t = (-2\beta h_t^2+\delta^2)\,dt + 2\delta h_t\,dW_{2t} \tag{S7}
$$

其中$h_t^2=v_t,\ h_t=\sqrt{v_t}$:
$$
dv_t = (\delta^2-2\beta v_t)\,dt + 2\delta\sqrt{v_t}\,dW_{2t} \tag{S8}
$$

对参数重命名:
$$
\kappa=2\beta,\qquad
\theta=\frac{\delta^2}{2\beta},\qquad
\sigma=2\delta \tag{S9}
$$

最终方差过程:
$$
dv_t = \kappa(\theta-v_t)\,dt + \sigma\sqrt{v_t}\,dW_{2t} \tag{S10}
$$

即为前述公式(2):
$$
dv_t = \kappa(\theta-v_t)\,dt + \sigma\sqrt{v_t}\,dW_{2t} \tag{2}
$$

# 费曼-卡茨定理Feynman-Kac-therorem
费曼-卡茨定理是链接偏微分方程PDE，与随机概率路径SDE的桥梁。  
假设一个风险中性的随机概率中的状态函数$X_t$，其可以是标的价格或方差等满足随机过程：
$$
dX_t = A(t, X_t)\,dt + B(t, X_t)\,dW_t\tag{S11}
$$
其中，$A(t,X_t)$为风险中性漂移率，也是状态函数$X_t$的期望轨迹。
假设衍生品价格$g(t, X_t)$，则根据伊藤引理，
$$
dg\,=\,
\left[
\frac{\partial g}{\partial t}
+ A_i \frac{\partial g}{\partial X_i}
+ \frac{1}{2}\Sigma_{ij}\frac{\partial^2 g}{\partial X_i \partial X_j}
\right] dt + \cdots\, dW_t\tag{S12}
$$
首先假定在一个无风险收益率r为0的风险中性随机概率过程中，则衍生品对时间$t$的一阶偏导为0。  
由此可得费曼-卡兹PDE：
$$
\frac{\partial g}{\partial t}
+ A_i \frac{\partial g}{\partial X_i}
+ \frac{1}{2}\Sigma_{ij}\frac{\partial^2 g}{\partial X_i \partial X_j}
\,=\,0\tag{S13}
$$
若考虑无风险收益率r，则考虑折现后衍生品价格$e^{-rt} g$，使用伊藤引理同样会得到一致的费曼-卡兹
PDE。  
注意，在费曼-卡兹PDE中，$\frac{\partial g}{\partial X_i}$前的因子即为$A_i$，即
状态函数$A$的风险中性漂移率。在Heston模型中，$\frac{\partial g}{\partial X_i}$前的因子
为$f(S,t,v)$。根据[资产定价基本定理](#资产定价基本定理)，$f(S,t,v)$即为$A_i$。

# 资产定价基本定理
无风险套利投资组合等价于风险中性世界定价。
因此我们可以通过两种方式获得等价的PDE：
1. 直接使用风险中性，则根据费曼-卡兹定理，$\frac{\partial g}{\partial v_i}$前的因子即为$A_i$，
其中$A_i$为方差的风险中性偏移率。
2. 使用Heston模型构建无风险套利投资组合后，标的价格被完全对冲，
$\frac{\partial g}{\partial v_i}$的因子为普适任意基于同一标的的衍生品的的函数$f(S,v,t)$。
由于两者等价，因此$f(S,v,t)$即为方差的风险中性偏移率。
