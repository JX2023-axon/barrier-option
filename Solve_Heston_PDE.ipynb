### Heston模型PDE公式
$$
\frac{\partial U}{\partial t}
+ \frac{1}{2}vS^2 \frac{\partial^2 U}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 U}{\partial S \partial v}
+ \frac{1}{2}\sigma^2 v \frac{\partial^2 U}{\partial v^2}
+ rS \frac{\partial U}{\partial S}
+ \kappa^*(\theta^* - v)\frac{\partial U}{\partial v}
- rU = 0
$$

可以转换成：
$$
\frac{\partial U}{\partial t} = AU
$$
在此基础上，在$U_{t_{i+1}}$已知的情况下，若获得偏导信息，则A和AU均可获得，从而获得$U_{t_i}$

### 偏导
泰勒展开  
$$
U_{i_{-1},j} = U_{i,j} - h_mU_s + \frac{h_m^2}{2}U_{ss} - \frac{h_m^2}{6}U_{sss} + ...
\\[1em]
U_{i_{+1},j} = U_{i,j} + h_pU_s + \frac{h_p^2}{2}U_{ss} + \frac{h_p^3}{6}U_{sss} + ...
\\[1em]
h_m = S_i - S_{i-1}， h_p = S{i+1} - S_i
$$

尝试获得
$$
U_s = a_{i-1}U_{i-1,j} + a_{i}U_{i,j} + a_{i+1}U_{i+1,j}
\\[1em]
U_s = (a_{i-1} + a_{i} + a{i+1})U_{i,j} + (h_pa_{i+1}-h_ma_{i-1})U_{s} + \frac{h_m^2a_{i-1}+h_p^2a_{i+1}}{2}U_{ss}
\\[1em]
a_{i-1} + a_{i} + a_{i+1} = 0
\\[1em]
h_pa_{i+1}-h_ma_{i-1} = 1
\\[1em]
h_m^2a_{i-1}+h_p^2a_{i+1} = 0
$$


解得
$$
a_{i+1} = \frac{h_m}{h_p(h_p+h_m)}
\\[1em]
a_{i-1} = -\frac{h_p}{h_m(h_p+h_m)}
\\[1em]
a_{i} = \frac{h_p-h_m}{h_ph_m}
$$

矩阵化
$$
D_S = \begin{bmatrix}
* & * & 0 & \dots & \dots & 0 \\
* & * & * & \dots & \dots & 0 \\
0 & * & * & \ddots &  & \vdots \\
\vdots & \dots & \ddots & \ddots & * & 0 \\
\vdots & \dots &  & * & * & * \\
0 & \dots & \dots & 0 & * & *
\end{bmatrix}
$$
其中，三个系数分别为$a_{i-1},a_{i},a_{i+1}$，在边界上，
$$
D_S[0,0] = -\frac{1}{S_1-S_0}, D_S[0,1] = \frac{1}{S_1-S_0},
\\[1em]
D_S[N_S-1,N_S-2] = -\frac{1}{S_{N_{S}-1}-S_{N_{S}-2}}, D_s[N_S-1,N_S-1] = \frac{1}{S_{N_S-1}-S_{N_S-2}}
$$

$$
U_{Sv} = \begin{bmatrix}
U_{S_1v_1} & U_{S_1v_2} & U_{S_1v_3} & \dots & \dots & U_{S_1v_{N_v}} \\
U_{S_2v_1} & U_{S_2v_2} & U_{S_2v_3} & \dots & \dots & U_{S_2v_{N_v}} \\
U_{S_3v_1} & U_{S_3v_2} & U_{S_3v_3} & \ddots &  & \vdots \\
\vdots & \dots & \dots & U_{S_{N_s-2}v_{N_v-2}} & U_{S_{N_s-1}v_{N_v-2}} & U_{S_{N_s-2}v_{N_v}} \\
\vdots & \dots &  \dots & U_{S_{N_s-1}v_{N_v-2}} & U_{S_{N_s-1}v_{N_v-1}} & U_{S_{N_s-1}v_{N_v}} \\
U_{S_{N_s}v_1} & \dots & \dots & U_{S_{N_s}v_{N_v-2}} & U_{S_{N_s}v_{N_v-1}} & U_{S_{N_s}v_{N_v}}
\end{bmatrix}

因此
$$
U_S = D_S U_{Sv}
\\[1em]
U_v = U_{Sv} D_v^T
$$

**二阶偏导**  
尝试获得
$$
U_{ss} = b_{i-1}U_{i-1,j} + b_{i}U_{i,j} + b_{i+1}U_{i+1,j}
\\[1em]
U_{ss} = (b_{i-1} + b_{i} + b{i+1})U_{i,j} + (h_pb_{i+1}-h_mb_{i-1})U_{s} + \frac{h_m^2b_{i-1}+h_p^2b_{i+1}}{2}U_{ss}
\\[1em]
b_{i-1} + b_{i} + b_{i+1} = 0
\\[1em]
h_pb_{i+1}-h_mb_{i-1} = 0
\\[1em]
h_m^2b_{i-1}+h_p^2b_{i+1} = 2
$$


解得
$$
b_{i+1} = \frac{2}{h_p(h_p+h_m)}
\\[1em]
b_{i-1} = \frac{2}{h_m(h_p+h_m)}
\\[1em]
b_{i} = -\frac{2}{h_mh_p}

同样矩阵化，且将边界处理为0：
$$
D_{SS} = \begin{bmatrix}
0 & 0 & 0 & \dots & \dots & 0 \\
* & * & * & \dots & \dots & 0 \\
0 & * & * & \ddots &  & \vdots \\
\vdots & \dots & \ddots & \ddots & * & 0 \\
\vdots & \dots &  & * & * & * \\
0 & \dots & \dots & 0 & 0 & 0
\end{bmatrix}
\\[1em]
U_{SS} = D_{SS}U_{Sv}
\\[1em]
U_{vv} = U_{Sv}D_{vv}^T
$$


**二阶混合偏导**
$$
U_{Sv} = U_S D_v^T = D_S U_{Sv} D_v^T
$$

### 系数矩阵
重新回到PDE本身
$$
\frac{\partial U}{\partial t}
+ \frac{1}{2}vS^2 \frac{\partial^2 U}{\partial S^2}
+ \rho \sigma v S \frac{\partial^2 U}{\partial S \partial v}
+ \frac{1}{2}\sigma^2 v \frac{\partial^2 U}{\partial v^2}
+ rS \frac{\partial U}{\partial S}
+ \kappa^*(\theta^* - v)\frac{\partial U}{\partial v}
- rU = 0
$$

$$
\frac{\partial U}{\partial t} = 
- \rho \sigma v S \frac{\partial^2 U}{\partial S \partial v}
- \frac{1}{2}vS^2 \frac{\partial^2 U}{\partial S^2}
- rS \frac{\partial U}{\partial S}
+ rU
- \frac{1}{2}\sigma^2 v \frac{\partial^2 U}{\partial v^2}
- \kappa^*(\theta^* - v)\frac{\partial U}{\partial v}
$$

因为此前，我们已经获得U对S,v的偏导均可以表达成D*U，由此有
$$
\frac{\partial U}{\partial t} = \mathcal{L}[U]
$$
因为此处我们从终末值向前推，因此更恰当的：
$$
\frac{\partial U}{\partial \tau} = \mathcal{L}[U]
\\[1em]
\tau = T - t
$$
通过PDE，期权价格U虽时间变化的$\frac{\partial U}{\partial \tau}$可以通过空间项$\mathcal{L}[U]$解得。
若将矩阵$U_{S,v}$转化成向量$u \in \mathbb{R}^{N_s N_v}$，则
$$
\frac{\partial u}{\partial \tau} = Au
$$

| Operation on *U* | Vectorized as *Au* |
|---|---|
| $D_s \cdot U$ | $(I_v \otimes D_S)u$ |  
| $U \cdot D_v^T$ | $(D_v \otimes I_S)u$ |  
| $D_S \cdot U \cdot D_v^T$ | $(D_v \otimes D_S)u$ |   

其中，关于$D_S \cdot U \cdot D_v^T$项
$$
D_S \cdot U \cdot D_v^T = ((I_v \otimes D_S)u) \cdot D_v^T = (D_v \otimes I_S)((I_v \otimes D_S)u) = (D_v \otimes D_S)u
$$

通过以上操作可以将均值运算转换成$\frac{\partial u}{\partial \tau} = Au$。若A已知，则
$$
\frac{u_{n+1} - u_{n}}{\Delta t} = Au_{n+1}
\\[1em]
u_{n} = (I - \Delta t A)u_{n+1}
$$
其中$u_n$是已知的时间轴下的值，而$u_{n+1}$是下一时间轴待解的值。此处的隐式格式直接求解涉及解大型线性方程组。因此我们在以下使用交替方向隐式法（Alternative Direction Implicit）进行求解，可用Douglas格式或修正的Craig-Sneyd格式（MCS）

### Douglas格式
若将S、v混合偏导方向，S偏导方向、v偏导方向拆分成$A_0$，$A_1$，$A_2$，则有：
显式求解：
$$
u_{n+1} = u_{n} + A\Delta t u_{n} = u_{n} + (A_1 + A_2 + A_3)\Delta t u_{n}
$$
显式求解在数值求解中存在稳定性不足的缺陷，通常需要更细的网格分隔来确保稳定性。

假设我们有$Y_1 = u_{n+1}$，且
$$
Y_1 = u_n + \Delta tA_0u_n + \Delta tA_2u_n + \Delta tA_1Y_1
$$
此时，$A_0$与$A_2$方向为显式求解，而$A_1$方向则是隐式求解。

假设
$$
Y_0 = u_n + \Delta tA_0u_n + \Delta tA_1u_n + \Delta tA_2u_n\tag{1}
$$
则
$$
Y_1 = Y_0 + \Delta tA_1Y_1 - \Delta tA_1u_n\tag{2}
$$

类似的，
$$
Y_2 = Y_0 + \Delta t A_0 u_n + \Delta t A_1 Y_1 + \Delta t A_2 Y_2
$$
其中，$A_0$项仍为显式解，$A_1$与$A_2$项则为隐式解。  
又因为
$$
Y_2 = Y_0 + \Delta t A_0 u_n + \Delta t A_1 Y_1 + \Delta t A_2 u_n + \Delta t A_2 Y_2 - \Delta t A_2 u_n
\\[1em]
Y_2 = Y_1 + \Delta t A_2 Y_2 - \Delta t A_2 u_n\tag{3}
$$
方程(1)，(2)，(3)即为Douglas范式的在$A_1$,$A_2$方向全隐式的解析过程。

若将方程(2)，(3)转变为：
$$
Y_1 = Y_0 + \theta \Delta t(A_1Y_1 - A_1u_n)
\\[1em]
Y_2 = Y_1 + \theta \Delta t(A_2Y_2 - A_2u_n)
$$
其中,$0<\theta<1$，则有
$$
Y_1 = u_n + \Delta t A_0 u_n + \Delta t A_2 u_n + \theta \Delta t A_1Y_1 + (1-\theta) \Delta t A_1u_n
\\[1em]
Y_2 = u_n + \Delta t A_0 u_n + \theta \Delta t A_1Y_1 + (1-\theta) \Delta t A_1u_n + \theta \Delta t A_2Y_2 + (1 - \theta) \Delta t A_2 u_n
$$
$A_1$与$A_2$方向变为显式与隐式所交替的项。

在实际计算中，
$$
Y_0 = u_n + (A_1 + A_2 + A_3) \Delta t\tag{4}
$$
$$
Y_1(1 - \theta \Delta t A_1) = Y_0 + \theta \Delta t A_1 u_n\tag{5}
$$
$$
Y_2(1 - \theta \Delta t A_2) = Y_1 + \theta \Delta t A_2 u_n\tag{6}
$$
其中，公式(4)是显性计算，直接可以获得。  
公式(5)、(6)的右侧可以直接获得，而左侧需要先计算括号内的内容，设为$M_1$和$M_2$，有：
$$
M_1 = 1 - \theta \Delta t A_1
\\[1em]
M_2 = 1 - \theta \Delta t A_2
$$

进一步的，在解：
$$
M_1 Y_1 = rhs_1
\\[1em]
M_2 Y_2 = rhs_2
$$
使用稀疏LU分解splu：
$PM_1Q = LU$，其中L、U为上三角矩阵和下三角矩阵，P、Q为排列矩阵
有：
$$
M_1 = P^{-1} L U Q^{-1}
\\[1em]
P^{-1} L U Q^{-1} Y_1 = rhs_1
\\[1em]
LUQ^{-1} Y_1 = Prhs_1
$$
在其中$Prhs_1$可以直接获得，而先后解$L\cdot$，和$U\cdot$在计算上要容易很多。

在Douglas格式中，我们有：
$$
U'(t) = AU(t)
$$
又因为
$$
U(t_n + \Delta t) = U(t_n) + \Delta t U'(t_n) + \Delta t^2 U''(t_n) + ...
\\[1em]
U(t_n + \Delta t) = U(t_n) + \Delta t AU(t) + \Delta t^2 A^2U(t_n) + ...
$$
因为Douglas格式使用$A = A_0 + A_1 + A_2$分布求解，因此其单步误差为$O(\Delta t^2)$，全局误差则为$O(\Delta t)$。MCS格式在Douglas格式的基础上加入额外步骤进行修正。

### MCS格式
修正Craig-Sneyd格式求解如下：  
前三步与Douglas格式一致。  
在此时我们获得$Y_2$，在Douglas格式下，我们将其视为$u_{n+1}$的值。同时，我们知道：
$$
Y_2 - u_n = \Delta t u_n + O(\Delta t^2)
$$
此时，
$$
\theta\Delta tA_0(Y_2 - u_n) \approx \theta\Delta t^2A_0A u_n
\\[1em]
(\frac{1}{2}-\theta)\Delta tA(Y_2 - u_n) \approx (\frac{1}{2}-\theta)\Delta t^2A^2 u_n
$$

令
$$
\hat Y = Y_0 + \theta\Delta t A_0(Y_2 - u_n)
\\[1em]
\tilde Y_0 = \hat Y + (\frac{1}{2} - \theta)\Delta tA(Y_2 - u_n)
\\[1em]
\tilde Y_0 = u_n + \Delta tAu_n + \Delta t^2[\theta A_0A + (\frac{1}{2} - \theta)A^2]u_n + O(\Delta t^3)
\\[1em]
\tilde Y_0 = u_n + \Delta tAu_n + \Delta t^2[\frac{1}{2}A^2 + \theta(A_0 - A)A]u_n + O(\Delta t^3)
$$

若有：
$Y = \tilde Y + \Delta t^2\theta (A_1 + A_2)Au_n$，则
$$
Y = u_n + \Delta t A u_n + \frac{1}{2}\Delta t^2 A^2 u_n + O(\Delta t^3)
$$
此时Douglas格式中的二次项误差得到了消除，而最终能获得$O(\Delta t^2)$误差，在时间层面能够更快收敛。为获得Y，我们有：
$$
\tilde Y_1 = \tilde Y_0 + \Delta t^2 \theta A_1Au_n
\\[1em]
\tilde Y_2 = \tilde Y_1 + \Delta t^2 \theta A_2Au_n
$$  
$\tilde Y_2$就是我们要解的Y。

考虑
$$
\tilde Y_1 = \tilde Y_0 + \theta \Delta t A_1(\tilde Y_0 - u_n)
$$
因为
$$
\tilde Y_0 = u_n + \Delta t A u_n + O(\Delta t^2)
$$
所以
$$
\tilde Y_1 = \tilde Y_0 + \theta \Delta t^2 A_1 A u_n + O(\Delta t^3)
$$
对$\tilde Y_2$进行一样的处理后，满足我们对二次项的校正，和全局$O(\Delta t^2)$误差。但此处的方程为显式，其隐式为：
$$
\tilde Y_1 = \tilde Y_0 + \theta \Delta t A_1(\tilde Y_1 - u_n)
\\[1em]
\tilde Y_2 = \tilde Y_0 + \theta \Delta t A_2(\tilde Y_2 - u_n)
\\[1em]
(I - \theta \Delta t A_1)\tilde Y_1 = \tilde Y_0 - \theta \Delta t A_1 u_n
\\[1em]
(I - \theta \Delta t A_1)\tilde Y_2 = \tilde Y_1 - \theta \Delta t A_2 u_n
$$

汇总后我们有MCS格式下的整体流程：
1. 最初显式估计
$$
Y_0 = u_n + \Delta t A u_n
$$
2. S，v方向隐式修正，获得Douglas格式结果$Y_2$
$$
(I - \theta \Delta t A_1) Y_1 = Y_0 - \theta \Delta t A_1 u_n
\\[1em]
(I - \theta \Delta t A_2) Y_2 = Y_1 - \theta \Delta t A_2 u_n
$$
3. 使用新的估计值$Y_2$处理$\Delta t$二次项
$$
\hat Y_0 = Y_0 + \theta \Delta t A_0 (Y_2 - u_n)
\\[1em]
\tilde Y_0 = \hat Y_0 + (\frac{1}{2} - \theta)\Delta t A(Y_2 - u_n)
$$
4. S，v方向隐式修正消除$\Delta t$二次项误差
$$
(I - \theta \Delta t A_1)\tilde Y_1 = \tilde Y_0 - \theta \Delta t A_1 u_n
\\[1em]
(I - \theta \Delta t A_2)\tilde Y_2 = \tilde Y_1 - \theta \Delta t A_2 u_n
$$

### 边界条件
对于障碍期权，$S$方向上的边界条件为敲出，即期权价值为0，或期权deep-in-the-money，期权价值直接按期权的收益获得。  
$v$方向上的边界条件，当$v$处在上边界时，认为其价格不在受$v$影响。当$v$为0时，带v项变为0，PDE退化为：
$$
\frac{\partial U}{\partial t} + rS\frac{\partial U}{\partial S} + k\theta\frac{\partial U}{\partial v} - rU = 0
$$

在实际计算中，因为$v$处在边界，$\frac{\partial U}{\partial v}$直接计算$dv$与0之间的偏导。剩余的带S的偏导项通过隐式解得。
